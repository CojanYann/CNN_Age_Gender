```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import cv2
#from google.colab.patches import cv2_imsh

from google.colab import files
from PIL import Image

from keras.preprocessing.image import load_img
from tensorflow.keras.models import load_model

#Chargement des models
model160 = load_model('Model_age_gender160x160_30epochs.h5')
model128 = load_model('Model_age_gender_128x128_20epochs.h5')

#Chargement et normalisation de l'image
Image = cv2.imread('Photo_128.jpg', cv2.IMREAD_GRAYSCALE)
Image = Image/255
Image.shape
```

```python
#Dictionnaire homme/femme
gender_dict = {0:'Male', 1:'Female'}
#affichage des donnée réels
#print('Original gender: ', gender_dict[y_gender[Image], 'Original age : ', y_age[Image])
#affichage des prdictions du model
pred = model160.predict(Image.reshape(1, 128, 128, 1))
pred_gender = gender_dict[round(pred[0][0][0])]
pred_age = round(pred[1][0][0])
print("Predicted gender : ", pred_gender, "Predicted age : ", pred_age)
plt.imshow(Image.reshape(128, 128), cmap = 'gray')
```
